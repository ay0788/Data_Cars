# **DataCars Prediction of Prices**

***This notebook explores car price prediction using a linear regression model based on features like present price and kilometers driven, further visualized using various plots.***

## **Importing Libraries**

```python
 import numpy as np
 import pandas as pd
 from matplotlib import pyplot as plt
 from sklearn import linear_model
 from sklearn.metrics import  r2_score
```

## **Loading the Dataset**

```python
df= pd.read_csv("CarData.csv")
df
```

check missing values

```python
df.isnull().sum()
```

data cleaning

```python
df.dropna()
```

checking another time missing values

```python
df.isnull()
```

```python
x=df[['Present_Price', 'Kms_Driven']]
y=df['Selling_Price']
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(x,y)
```

## **Selling Price Prediction**

```python
price=int(input("enter the present price $:"))
kms=int(input("enter the kms:"))
```

```python
prediction=reg.predict([[price,kms]])
print("the predicted price is:",prediction)
```

Visualizing car selling prices using a line plot

The x-axis represents the index of the data, while the y-axis shows the selling price of each car

```python
df['Selling_Price'].plot(kind='line', figsize=(10, 8), title='Selling_Price')
plt.show()
```

Scatter plot illustrating the distribution of car Selling Prices in relation to their Present Prices

Each point represents a car, with transparency for better visualization

```python
df.plot(kind='scatter', x='Selling_Price', y='Present_Price', s=32, alpha=.8)
plt.show()
```

analyzing the relationship between a car's selling price and its present price

```python
r = df['Selling_Price'].corr(df['Present_Price'])
val="value of r : {}"
print(val.format(r))
```

R-squared; proportion of variance in selling price explained by present price

```python
r2 = r2_score(y, reg.predict(x))
print("r2 score : {} ".format(r2))
```

## Distribution of Present and Selling Prices of Cars

```python
plt.hist(df[['Present_Price','Selling_Price']], bins=10)
plt.show()
```

 ***This analysis demonstrates a positive correlation between present and selling prices, allowing for reasonable price predictions using a linear regression model and visualization techniques for better understanding the data.***
